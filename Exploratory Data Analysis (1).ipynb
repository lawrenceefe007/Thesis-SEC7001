```python
#plotting stock price over time

# Read the CSV data into a pandas DataFrame
df = pd.read_csv("C:\\Users\\lefen\\Downloads\\AAPL.csv", parse_dates=['Date'], dayfirst=True)

# Set the Date column as the index of the DataFrame
df.set_index('Date', inplace=True)

# Plot the data
plt.figure(figsize=(18,9))
plt.plot(df['Open'], label='Open')
plt.plot(df['Close'], label='Close')
plt.plot(df['High'], label='High')
plt.plot(df['Low'], label='Low')


plt.title('Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
#histogram of daily price changes

# Read the CSV data into a pandas DataFrame
df = pd.read_csv("C:\\Users\\lefen\\Downloads\\AAPL.csv", parse_dates=['Date'], dayfirst=True)

# Set the Date column as the index of the DataFrame
df.set_index('Date', inplace=True)

# Calculate the daily price change
df['Price Change'] = df['Close'] - df['Open']

# Plot the histogram
plt.figure(figsize=(14, 8))
plt.hist(df['Price Change'], bins=30, color='blue', edgecolor='black')
plt.title('Histogram of Daily Price Changes')
plt.xlabel('Price Change')
plt.ylabel('Frequency')
plt.show()
```

```python
#stock closing prices and moving averages


# Read the CSV data into a pandas DataFrame
df = pd.read_csv("C:\\Users\\lefen\\Downloads\\AAPL.csv", parse_dates=['Date'], dayfirst=True)

# Set the Date column as the index of the DataFrame
df.set_index('Date', inplace=True)

# Calculate the moving averages
df['7 Day Avg'] = df['Close'].rolling(window=7).mean()
df['30 Day Avg'] = df['Close'].rolling(window=30).mean()

# Plot the closing prices and moving averages
plt.figure(figsize=(18,9))
plt.plot(df['Close'], label='Close')
plt.plot(df['7 Day Avg'], label='7 Day Average', linestyle='--')
plt.plot(df['30 Day Avg'], label='30 Day Average', linestyle='--')

plt.title('Stock Closing Prices and Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
#monthly comparison of stock open and close prices


# Read the CSV data into a pandas DataFrame
df = pd.read_csv("C:\\Users\\lefen\\Downloads\\AAPL.csv", parse_dates=['Date'], dayfirst=True)

# Set the Date column as the index of the DataFrame
df.set_index('Date', inplace=True)

# Resample the data to a monthly frequency
monthly_df = df.resample('M').mean()

# Plot the monthly open and close prices
plt.figure(figsize=(14,7))
plt.plot(monthly_df['Open'], label='Open', marker='o' , color='red')
plt.plot(monthly_df['Close'], label='Close', marker='o' , color='green')

plt.title('Monthly Comparison of Stock Open and Close Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
#monthly comparison of stock high and low prices


# Read the CSV data into a pandas DataFrame
df = pd.read_csv("C:\\Users\\lefen\\Downloads\\AAPL.csv", parse_dates=['Date'], dayfirst=True)

# Set the Date column as the index of the DataFrame
df.set_index('Date', inplace=True)

# Resample the data to a monthly frequency
monthly_df = df.resample('M').mean()

# Plot the monthly high and low prices
plt.figure(figsize=(14,7))
plt.plot(monthly_df['High'], label='High', marker='o' , color='red')
plt.plot(monthly_df['Low'], label='Low', marker='o' , color='green')

plt.title('Monthly Comparison of Stock High and Low Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```
